```python
import dotenv
import os
dotenv.load_dotenv()
```

```python
from scapy.utils import RawPcapReader, PcapReader
from scapy.all import *

load_layer("inet")
load_layer("tls")
load_layer("http")
load_layer("dns")
```

```python
data_dir = os.getenv("DATA_DIR", './data')
```

```python
pcap_path = os.path.join(data_dir, 'Monday-WorkingHours.pcap')
pcap_path
```

```python
with PcapReader(pcap_path) as pcap:    
    first_pkt = next(pcap)
    first_pkt.show()
```

```python
c = 0

with RawPcapReader(pcap_path) as pcap:
    for pkt in pcap:
        c += 1

c
```
